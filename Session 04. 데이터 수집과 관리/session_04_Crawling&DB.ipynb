```python
# 순위 변동 칼럼이 계산하기 불편하니 바꿔주자.

def change_rank(rank):
    if rank == "동일":
        rank = 0
    elif rank[-2:] == "하락":
        rank = int(rank[:-2]) * -1
    elif rank[-2:] == "상승":
        rank = int(rank[:-2]) * 1
    return rank

daily_boxoffice['순위 변동'] = daily_boxoffice['순위 변동'].map(lambda x: change_rank(x))

        
daily_boxoffice # 이렇게 정리한 데이터를 어디다가 저장해야할까 ?
```
